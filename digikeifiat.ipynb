## Reading xlsx file using pandas.read_excel function

```python
import pandas as pd
keifiat_dataframe = pd.read_excel('digi_keifiat.xlsx')
```

```python
keifiat_dataframe
```

```python
keifiat_dataframe.shape
```

```python
# showing a statistical summary of the dataframe
keifiat_dataframe.describe()
```

```python
# Accessing all rows of mobile phone product
mobile_dataframe = keifiat_dataframe.loc[(keifiat_dataframe['title_en'] == ('MO'))]
mobile_dataframe
```

```python
# Accessing all the mobile phones with word "اپل" in their product_title
apple_dataframe = mobile_dataframe.loc[(mobile_dataframe['product_title'].str.contains('اپل'))]
apple_dataframe
```

```python
# sorting the rows by number of likes(descending)
apple_dataframe.sort_values('likes', ascending=False)
```

```python
# add a like to dislike ratio column
apple_dataframe['like_ratio'] = apple_dataframe['likes'] / apple_dataframe['dislikes']
```

```python
apple_dataframe[['product_title','like_ratio']]
```

```python
apple_dataframe[['product_title','like_ratio']].to_excel('apple_phones.xlsx', index=False)
```
